# Focus Your Audience's Attention

You should focus your audience's attention on the most important parts of the previous visuals that you had drawn for your data visualization project. In particular, you should do the following:
- Focus on a few visuals that you had drawn in previous sessions.
- Focus your audience's attention on your story by adding preattentive attributes to the visuals.
- Critically evaluate your design decisions. Why does adding these preattentive attributes improve your visuals?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('/content/health&lifestyle.zip')
df.head()
```

```python
sns.histplot(data=df, x='Age', kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# decluttered graph
sns.histplot(data=df, x='Age', kde=True, bins=20, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Comparing to the previous graph done by Brian Risk(https://www.kaggle.com/code/devraai/diabetes-prediction-and-lifestyle-analysis), a more effective color is used to visualise the graph. The 'bins' paramter is reduced from 30 to 20 to show the bars more wider. The 'alpha' paramter and linestyle reduced the grid lines and width.

The decluttered histogram clearly showed the overall age distribution of patients. However, it didnâ€™t immediately direct the viewerâ€™s attention to the most relevant insight â€” the age group with the highest frequency.

```python
# adding preattentive attributes to the visuals
plt.figure(figsize=(8,5))

sns.histplot(data=df, x='Age', kde=True, bins=20, color='lightgrey')

# the 50â€“60 age group (most common)
highlight = df[(df['Age'] >= 50) & (df['Age'] <= 60)]
sns.histplot(data=highlight, x='Age', kde=True, bins=20, color='green')

plt.axvspan(50, 60, color='green', alpha=0.2, label='Focus: Ages 50â€“60')
plt.title('Most Patients Are in Their 50s')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
sns.despine()
plt.show()

```

Impact on Visual Clarity

These preattentive attributes help the audience instantly recognize the central insight without reading all axis labels or interpreting density curves. The chart now clearly communicates that most patients in the dataset are in their 50s, making it more focused and story-driven.

| **Aspect**           | **Before (Decluttered Visual)**                                                                      | **After (With Preattentive Attributes)**                                                                |
| -------------------- | ---------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------- |
| **Color Usage**      | All bars shown in uniform green; no specific focus area.                                             | Age range 50â€“60 highlighted in green, while other bars are light grey â€” creates immediate visual focus. |
| **Message Focus**    | Title simply stated â€œAge Distribution,â€ requiring viewers to interpret the main pattern themselves.  | Title changed to â€œMost Patients Are in Their 50s,â€ clearly communicating the key insight upfront.       |
| **Visual Elements**  | Smooth density curve and gridlines retained for readability, but no emphasis on specific data range. | Added shaded region (50â€“60) and color contrast to emphasize the most populated age group.               |
| **Viewer Attention** | Viewer needs to scan across all bins to find where most patients lie.                                | The eye is instantly drawn to the highlighted range, reducing cognitive load and improving clarity.     |
| **Overall Impact**   | Clean and readable, but descriptive rather than insightful.                                          | Insight-driven and focused; helps the audience understand the main trend immediately.                   |
