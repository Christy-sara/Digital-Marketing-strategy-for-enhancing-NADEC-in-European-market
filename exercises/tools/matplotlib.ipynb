### Plot with Different Interfaces
Plot the function $ y = sin(x) * cos(x) $ using both MATLAB-style and object-oriented interfaces. Decorate the plots properly and save them as png files.

```python
x = np.linspace(0, 2*np.pi, 100)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x, y1, label='sin(x)', color='green')
plt.plot(x, y2, label='cos(x)', color='red')
plt.title("Sine and Cosine Waves")
plt.xlabel("x (radians)")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### Plot Multiple Functions
Draw three plots in a row. The first one show draw the function $ y = log_2x $. The second one show draw the function $ y = \sqrt{x} $. The third one should draw both the functions together. Decorate the plots properly.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 10, 100)
y1 = np.log2(x)
y2 = np.sqrt(x)

plt.figure(figsize=(15, 4))

#First plot: y = log2(x)
plt.subplot(1, 3, 1)
plt.plot(x, y1, color='teal')
plt.title("y = log(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

#Second plot: y = sqrt(x)
plt.subplot(1, 3, 2)
plt.plot(x, y2, color='orange')
plt.title("y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

#Third plot: both functions together
plt.subplot(1, 3, 3)
plt.plot(x, y1, label='log(x)', color='teal')
plt.plot(x, y2, label='x', color='orange')
plt.title("Both Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Plot a Bar Chart with Error Bars
Plot a bar chart with error bars for each feature of the iris dataset. The height of each bar should represent the mean of the feature and the error bar should represent the standard deviation of the feature. Decorate the plot properly.

```python
import sklearn.datasets
iris = sklearn.datasets.load_iris().data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn.datasets

iris = sklearn.datasets.load_iris().data

feature_names = ['sepal length', 'sepal width', 'petal length', 'petal width']

means = np.mean(iris, axis=0)
stds = np.std(iris, axis=0)


plt.figure(figsize=(8, 5))
plt.bar(feature_names, means, yerr=stds, capsize=8, color='lightblue', edgecolor='black')

plt.title("Mean and Standard Deviation of Iris Features", fontsize=14, fontweight='bold')
plt.xlabel("Features", fontsize=12)
plt.ylabel("Mean Value", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Draw Subplots
Plot a 2x2 grid. Each subplot of this grid should draw the histogram of one feature of the iris dataset. Decorate the plot properly.

```python
import sklearn.datasets
iris = sklearn.datasets.load_iris().data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn.datasets


iris = sklearn.datasets.load_iris().data

feature_names = ['sepal length', 'sepal width', 'petal length', 'petal width']


fig, axes = plt.subplots(2, 2, figsize=(10, 8))


axes = axes.flatten()

for i in range(4):
    axes[i].hist(iris[:, i], bins=15, color='skyblue', edgecolor='black')
    axes[i].set_title(feature_names[i], fontsize=12, fontweight='bold')
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Frequency")
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Draw a 3D Plot
Visualize the data points of the iris dataset in a 3D space using the first three features. Decorate the figure properly.

```python
import sklearn.datasets
iris = sklearn.datasets.load_iris().data
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # enables 3D plotting
import sklearn.datasets
import numpy as np

iris = sklearn.datasets.load_iris()
data = iris.data
feature_names = iris.feature_names


fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x = data[:, 0]
y = data[:, 1]
z = data[:, 2]


ax.scatter(x, y, z, c='skyblue', edgecolor='black', s=50)


ax.set_xlabel(feature_names[0], fontsize=12)
ax.set_ylabel(feature_names[1], fontsize=12)
ax.set_zlabel(feature_names[2], fontsize=12)
ax.set_title("3D Scatter Plot of Iris Dataset", fontsize=14, fontweight='bold')

plt.show()
```
